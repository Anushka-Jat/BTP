```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/home/soham/Desktop/GitHub/BTP/data/time_data/no_of_days/Sensor 3_no of days_analyte 3.xlsx')

df_long = df.melt(id_vars=['Day'], value_vars=[f'Dil.{i}' for i in range(1, 11)],
                  var_name='Dilution', value_name='Sensor_Reading')

print(df_long.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_long, x='Dilution', y='Sensor_Reading', width=0.5)

plt.title('Sensor Readings Distribution Across Dilutions')
plt.xlabel('Dilution')
plt.ylabel('Sensor Reading')

plt.show()
```
